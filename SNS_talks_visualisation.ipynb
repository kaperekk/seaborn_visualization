```python

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df = pd.read_csv('/kaggle/input/ted-talks/data.csv')
```

```python
df
```

## Most Viewed Ted Talks

```python
ten_most_viewed_talks = df.sort_values("views",ascending=False).head(10)
```

```python
ten_most_viewed_talks[['author','views']]
```

```python
sns.set_theme()
plt.figure(figsize=(15,8))
ax = sns.barplot(x="author", y="views", data=ten_most_viewed_talks[['author','views']].head(10))
ax.tick_params(axis='x', rotation=30)
```

## Most Liked Ted Talks

```python
ten_most_liked_talks = df.sort_values("likes",ascending=False).head(10)
```

```python
ten_most_liked_talks[['author','likes']]
```

```python
sns.set_theme()
plt.figure(figsize=(15,8))
ax = sns.barplot(x="author", y="likes", data=ten_most_liked_talks[['author','likes']].head(10))
ax.tick_params(axis='x', rotation=30)
```
